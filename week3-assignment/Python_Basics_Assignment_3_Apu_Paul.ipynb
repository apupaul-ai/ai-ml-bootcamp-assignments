```python
# -------------------------------------------------
# Question 1: Enumerate with List
# This program prints each fruit in the list along
# with its index number, using enumerate().
# -------------------------------------------------

fruits = ["apple", "banana", "cherry", "mango"]

# enumerate() gives both the index and the value
# of each item in the list, one pair at a time
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
# -------------------------------------------------
# Question 2: Zip with Two Lists
# This program pairs up student names with their
# scores using zip(), then prints them together.
# -------------------------------------------------

students = ["Rahim", "Karim", "Ayesha"]
scores = [85, 90, 78]

# zip() combines the two lists element by element:
# (students[0], scores[0]), (students[1], scores[1]), ...
for student, score in zip(students, scores):
    print(f"{student} scored {score}")
```

```python
# -------------------------------------------------
# Question 3: Tuple and Loop
# This program creates a tuple of 5 subjects and
# prints each one using a for loop.
# -------------------------------------------------

# Tuples are created using round brackets ()
# and cannot be changed after creation
subjects = ("Math", "Physics", "Chemistry", "Biology", "English")

# Loop through the tuple and print each subject
for subject in subjects:
    print(subject)
```

```python
# -------------------------------------------------
# Question 4: Set Operations
# This program finds the common elements between two
# sets, and the elements unique to set1.
# -------------------------------------------------

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

# 1. Common elements: intersection (&) gives items
#    present in BOTH sets
common_elements = set1 & set2

# 2. Unique elements of set1: difference (-) gives
#    items in set1 that are NOT in set2
unique_to_set1 = set1 - set2

print(f"Common elements: {common_elements}")
print(f"Unique elements of set1: {unique_to_set1}")
```

```python
# -------------------------------------------------
# Question 5: Dictionary with Enumerate
# This program prints the subject number, subject
# name, and marks using enumerate() on a dictionary.
# -------------------------------------------------

marks = {
    "Math": 80,
    "Physics": 75,
    "Chemistry": 85
}

# marks.items() gives (key, value) pairs, e.g. ("Math", 80)
# enumerate(..., start=1) numbers them starting from 1
# instead of the default 0
for number, (subject, mark) in enumerate(marks.items(), start=1):
    print(f"{number} {subject}: {mark}")
```
