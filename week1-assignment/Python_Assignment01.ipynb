```python
# -------------------------------------------------
# Q1. Simple Calculator Program
# This program takes two numbers from the user and
# performs addition, subtraction, multiplication,
# and division on them.
# -------------------------------------------------

# --- A. Input & Setup ---

# Ask the user to input two numbers (as text first)
num1_input = input("Enter the first number: ")
num2_input = input("Enter the second number: ")

# Convert both inputs into float so decimals are supported
num1 = float(num1_input)
num2 = float(num2_input)

# --- B. Calculations ---

# Addition
addition = num1 + num2

# Subtraction
subtraction = num1 - num2

# Multiplication
multiplication = num1 * num2

# Division (check for zero to avoid a crash)
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (division by zero)"

# --- Clearly labeled outputs ---
print("\n----- Calculator Results -----")
print(f"Addition       : {num1} + {num2} = {addition}")
print(f"Subtraction    : {num1} - {num2} = {subtraction}")
print(f"Multiplication : {num1} * {num2} = {multiplication}")
print(f"Division       : {num1} / {num2} = {division}")
print("-------------------------------")
```

```python
# -------------------------------------------------
# Q2. User Introduction Program
# This program collects the user's name, city, and
# favourite hobby, then prints an introduction sentence.
# -------------------------------------------------

# --- A. Inputs ---

name = input("Enter your name: ")
city = input("Enter your city: ")
hobby = input("Enter your favourite hobby: ")

# --- B. Output ---

# Using an f-string to build a clear, readable sentence
introduction = f"Hello {name}! You live in {city} and you enjoy {hobby}."

print("\n----- Introduction -----")
print(introduction)
print("-------------------------")
```

```python
# -------------------------------------------------
# Q3. Temperature Converter (Celsius to Fahrenheit)
# This program converts a temperature given in
# Celsius into Fahrenheit using the standard formula.
# -------------------------------------------------

# --- A. Input ---

# Take temperature in Celsius from the user
celsius_input = input("Enter temperature in Celsius: ")

# Convert to float to allow decimal values
celsius = float(celsius_input)

# --- B. Calculation ---

# Formula: F = (C * 9/5) + 32
fahrenheit = (celsius * 9 / 5) + 32

# --- C. Output ---

print(f"Temperature in Fahrenheit is: {fahrenheit}°F")
```

```python
# -------------------------------------------------
# Q4. Billing Program
# This program calculates the total bill for an item
# based on its price and quantity.
# -------------------------------------------------

# --- A. Inputs ---

item_name = input("Enter item name: ")
price = float(input("Enter price per unit: "))
quantity = int(input("Enter quantity: "))

# --- B. Calculation ---

# Total = price × quantity
total = price * quantity

# --- C. Output ---

print(f"\nTotal Bill for {item_name} = {total}")
```

```python
# -------------------------------------------------
# Q5. Student Information Card
# This program takes student details and prints a
# neatly formatted Student ID Card.
# -------------------------------------------------

# --- A. Input ---

student_name = input("Enter Student Name: ")
class_section = input("Enter Class and Section (e.g. 8B): ")
school_name = input("Enter School Name: ")

# --- B. Formatting ---

# Using a multi-line f-string for clean, aligned output
id_card = f"""
----- STUDENT ID CARD -----
Name: {student_name}
Class: {class_section}
School: {school_name}
----------------------------
"""

print(id_card)
```
