```python
#-----------------------------------------------
# Question 1: Loop Basics (for loop)
# This program prints numbers from 1 to 10 using a
# for loop.
#-------------------------------------------------

# range(1, 11) generates numbers from 1 up to 10
# (11 is excluded, that's why we write 11 and not 10)


for number in range(1, 11):
    print(number)
```

```python
#---------------------------------------------------
# Question 2: Loop with Condition
# This program prints only the even numbers between
# 1 and 20 using a for loop.
# -------------------------------------------------

# range(1, 21) generates numbers from 1 to 20
for number in range(1, 21):
    # A number is even if it is exactly divisible by 2
    # (no remainder), checked using the modulo operator %
    if number % 2 == 0:
        print(number)
```

```python
# -------------------------------------------------
# Question 3: String Manipulation
# This program takes a string from the user and prints:
# 1. The string in uppercase
# 2. The length of the string
# 3. The string reversed
# -------------------------------------------------

# Take a string input from the user
text = input("Enter a string: ")

# 1. Convert the string to uppercase using .upper()
uppercase_text = text.upper()

# 2. Find the length of the string using len()
text_length = len(text)

# 3. Reverse the string using slicing [::-1]
# ( [::-1] means "go through the string backwards" )
reversed_text = text[::-1]

# Print all results clearly
print(f"Uppercase: {uppercase_text}")
print(f"Length: {text_length}")
print(f"Reversed: {reversed_text}")
```

```python
# -------------------------------------------------
# Question 4: String Processing with Loop
# This program takes a string from the user and counts
# how many vowels (a, e, i, o, u) are in it.
# -------------------------------------------------

# Take a string input from the user
text = input("Enter a string: ")

# Convert to lowercase first, so we catch both
# 'A' and 'a' as vowels without writing extra checks
text_lower = text.lower()

# Define what counts as a vowel
vowels = "aeiou"

# Counter to keep track of how many vowels we find
vowel_count = 0

# Loop through each character in the string one by one
for character in text_lower:
    # Check if the current character is a vowel
    if character in vowels:
        vowel_count += 1  # same as vowel_count = vowel_count + 1

# Print the final result
print(f"Number of vowels in the string: {vowel_count}")
```

```python
# -------------------------------------------------
# Question 5: List Comprehension
# Given a list of numbers, create a new list containing
# only the numbers greater than 40, using list
# comprehension.
# -------------------------------------------------

numbers = [10, 25, 30, 45, 50, 65, 70]

# List comprehension breakdown:
#   [expression for item in list if condition]
#   -> take "num" from "numbers", keep it only if num > 40
greater_than_40 = [num for num in numbers if num > 40]

# Print the new list
print(f"Original list: {numbers}")
print(f"Numbers greater than 40: {greater_than_40}")
```
